# Q1 • Plot data for all periods to the low value of the stock and study this graph in relation to change in the open value.


```python
plt.figure(figsize=(8,8))
d1['Open'].plot(label='Open')
d1['Low'].plot(label='Low')
plt.xlabel("Date")
plt.ylabel("Prices")
plt.title("Stock Prices")
plt.legend()
plt.show()
```

# Q2 • Analyze the price change within a day using standard deviation.



```python
add = (d1['Open'] + d1['Low'] + d1['High'] + d1['Close'])
mean=add/4
mo = (d1['Open']-mean)*(d1['Open']-mean)
ml = (d1['Low']-mean)*(d1['Low']-mean)
mh = (d1['High']-mean)*(d1['High']-mean)
mc = (d1['Close']-mean)*(d1['Close']-mean)
var = (mo+ml+mh+mc)/4
std = var**0.5

std.plot(figsize=(20,8),color="green")
plt.xlabel("Date")
plt.ylabel("Deviation")
plt.title("{Price change within a day")
plt.show()
```

#Q3 • Use a suitable plot to show the change in median of open, high, low and close in each quartile as per the volume.



```python
med = (d1['High'], d1['Low'], d1['Open'], d1['Close'])
data = pd.DataFrame(med)
plt.plot(data.median(),d1['Volume'])
plt.xlabel("Median")
plt.ylabel("Volume")
plt.title("Stock")
plt.yticks([360100, 1007500, 2206300, 3354800, 4937300])
plt.show()
```

Q4 Given that the difference between high and low is below the average and that of open and close is below the average, find the mean volume and compare it to the original value.

```python
mean = (d1['Volume'].sum()/d1['Date'].count()+ d1['Volume']-d1['Volume'])
mean.plot(figsize=(15,7),label='Mean',color='black')
plt.plot(d1['Volume'])
plt.xlabel('No of Days')
plt.ylabel('Volume')
plt.title('Comparison between mean and original volume')
plt.yticks([139100,360100,728668,1507500,2206300,3354800,4937300])
plt.legend()
plt.show()
```

Q5  Use a suitable plot to study the changes in stock trading with respect to time, also plot the changes of difference between open and close to study the fluctuations and relationship between the two.

```python
d1 ['Stock Trading'].plot(label='Stock Trading',figsize=(20,8),color='cyan')
(d1['Open']-d1['Close']).plot(label='Open-Close',figsize=(20,8),color='red')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Stock")
plt.show()
```

# Q6  Find the percentage change in median for the stock trading if the difference between high and low is below the average and that of open and close is above the average.
